```python
# mixed datatypes in isOriginalTitle column
df['isOriginalTitle'].value_counts()
```

```python
# updating data type to int
df.loc[df['isOriginalTitle'] == '1', 'isOriginalTitle'] = 1
df.loc[df['isOriginalTitle'] == '0', 'isOriginalTitle'] = 0
df.loc[df['isOriginalTitle'] == '\\N', 'isOriginalTitle'] = np.nan
```

```python
df['isOriginalTitle'].value_counts()
```

```python
originalTitles = pd.DataFrame()
originalTitles = df[df.isOriginalTitle == 1]
```
